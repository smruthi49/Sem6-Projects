```python
from AES import encrypt,decrypt

import numpy as np
import ast
import matplotlib.pyplot as plt
import cv2
import PIL
from PIL import Image
import io, os
```

```python
def show_img(img, title= '',ax = None):
    if ax is None:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(title);
    else :
        ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        ax.axis('off')
        ax.set_title(title);
```

```python
# Read image file and convert to bytes
def read_image(filename):
    with open(filename, "rb") as img_file:
        return img_file.read()

# Save bytes to a file
def save_bytes(filename, data):
    with open(filename, "wb") as file:
        file.write(data)

# Load bytes from a file
def load_bytes(filename):
    with open(filename, "rb") as file:
        return file.read()

# Convert bytes to image
def bytes_to_image(data):
    image = Image.open(io.BytesIO(data))
    return image

# Save image to file
def save_image(filename, image):
    image.save(filename)
```

```python
key = "hello world"
key.encode('utf-8')

image_data = read_image("brain.jpg")

img = bytes_to_image(image_data)
show_img(np.array(img), "Original Image")
```

```python
encrypted_data = encrypt(key, image_data)

save_bytes("encrypted_image.dat", encrypted_data)

decrypted_data = decrypt(key, encrypted_data)

decrypted_image = bytes_to_image(decrypted_data)

save_image("decrypted_image.jpg", decrypted_image)
```
